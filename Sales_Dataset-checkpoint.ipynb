# **RFM Model**

```python
from datetime import timedelta
#Create snapshot date
snapshot_date = data['Order_Date'].max() + timedelta(days=1)
print(snapshot_date)
# Grouping by CustomerID
data_new= data.groupby(['Customer_ID']).agg({
        'Order_Date': lambda x: (snapshot_date - x.max()).days,
        'Product_ID': 'count',
        'Amount_Spent': 'sum'})
```

```python
# Rename the columns 
data_new.rename(columns={'Order_Date': 'Recency',
                         'Product_ID': 'Frequency',
                         'Amount_Spent': 'Monetary Value'}, inplace=True)
```

```python
data_new.head()
```

```python
data_new.shape
```

```python
# Plot RFM distributions
plt.figure(figsize=(12,10))
# Plot distribution of R
plt.subplot(3, 1, 1); sns.distplot(data_new['Recency'])
# Plot distribution of F
plt.subplot(3, 1, 2); sns.distplot(data_new['Frequency'])
# Plot distribution of M
plt.subplot(3, 1, 3); sns.distplot(data_new['Monetary Value'])
# Show the plot
plt.show()
```

```python
# --Calculate R and F groups--
# Create labels for Recency and Frequency
r_labels = range(4, 0, -1); f_labels = range(1, 5)
# Assign these labels to 4 equal percentile groups 
r_groups = pd.qcut(data_new['Recency'], q=4, labels=r_labels)
# Assign these labels to 4 equal percentile groups 
f_groups = pd.qcut(data_new['Frequency'], q=4, labels=f_labels)
# Create new columns R and F 
data_new= data_new.assign(R = r_groups.values, F = f_groups.values)
data_new.head()
```

```python
# Create labels for MonetaryValue
m_labels = range(1, 5)
# Assign these labels to three equal percentile groups 
m_groups = pd.qcut(data_new['Monetary Value'], q=4, labels=m_labels)
# Create new column M
data_new = data_new.assign(M = m_groups.values)
```

```python
# Concat RFM quartile values to create RFM Segments
def join_rfm(x): return str(x['R']) + str(x['F']) + str(x['M'])
data_new['RFM_Segment_Concat'] = data_new.apply(join_rfm, axis=1)
rfm = data_new
rfm.head()
```

```python
# Count num of unique segments
rfm_count_unique = rfm.groupby('RFM_Segment_Concat')['RFM_Segment_Concat'].nunique()
print(rfm_count_unique.sum())
```

```python
# Calculate RFM_Score
rfm['RFM_Score'] = rfm[['R','F','M']].sum(axis=1)
print(rfm['RFM_Score'].head())
```

```python
# Define rfm_level function
def rfm_level(df):
    if df['RFM_Score'] >= 9:
        return 'Can\'t Loose Them'
    elif ((df['RFM_Score'] >= 8) and (df['RFM_Score'] < 9)):
        return 'Champions'
    elif ((df['RFM_Score'] >= 7) and (df['RFM_Score'] < 8)):
        return 'Loyal'
    elif ((df['RFM_Score'] >= 6) and (df['RFM_Score'] < 7)):
        return 'Potential'
    elif ((df['RFM_Score'] >= 5) and (df['RFM_Score'] < 6)):
        return 'Promising'
    elif ((df['RFM_Score'] >= 4) and (df['RFM_Score'] < 5)):
        return 'Needs Attention'
    else:
        return 'Require Activation'
# Create a new variable RFM_Level
rfm['RFM_Level'] = rfm.apply(rfm_level, axis=1)
# Print the header with top 5 rows to the console
rfm.head()
```

```python
# Calculate average values for each RFM_Level, and return a size of each segment 
rfm_level_agg = rfm.groupby('RFM_Level').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary Value': ['mean', 'count']
}).round(1)
# Print the aggregated dataset
print(rfm_level_agg)
```

# **K Means Clustering**

```python
data_new.describe()
```

```python
from scipy import stats
customers_fix = pd.DataFrame()
customers_fix["Recency"] = stats.boxcox(data_new['Recency'])[0]
customers_fix["Frequency"] = stats.boxcox(data_new['Frequency'])[0]
customers_fix["MonetaryValue"] = stats.boxcox(data_new['Monetary Value'])[0]
customers_fix.tail()
```

```python
# Import library
from sklearn.preprocessing import StandardScaler
# Initialize the Object
scaler = StandardScaler()
# Fit and Transform The Data
scaler.fit(customers_fix)
customers_normalized = scaler.transform(customers_fix)
# Assert that it has mean 0 and variance 1
print(customers_normalized.mean(axis = 0).round(2)) # [0. -0. 0.]
print(customers_normalized.std(axis = 0).round(2)) # [1. 1. 1.]
```

```python
from sklearn.cluster import KMeans
sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(customers_normalized)
    sse[k] = kmeans.inertia_ # SSE to closest cluster centroid
plt.title('The Elbow Method')
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
model = KMeans(n_clusters=3, random_state=42)
model.fit(customers_normalized)
model.labels_.shape
```

```python
data_new["Cluster"] = model.labels_
data_new.groupby('Cluster').agg({
    'Recency':'mean',
    'Frequency':'mean',
    'Monetary Value':['mean', 'count']}).round(2)
```

```python
# Create the dataframe
df_normalized = pd.DataFrame(customers_normalized, columns=['Recency', 'Frequency', 'Monetary Value'])
df_normalized['ID'] = data_new.index
df_normalized['Cluster'] = model.labels_
# Melt The Data
df_nor_melt = pd.melt(df_normalized.reset_index(),
                      id_vars=['ID', 'Cluster'],
                      value_vars=['Recency','Frequency','Monetary Value'],
                      var_name='Attribute',
                      value_name='Value')
df_nor_melt.head()
# Visualize it
sns.lineplot('Attribute', 'Value', hue='Cluster', data=df_nor_melt)
```

```python
cluster_avg = data_new.groupby('Cluster').mean()
population_avg = data_new.mean()
relative_imp = cluster_avg / population_avg - 1
relative_imp
```
